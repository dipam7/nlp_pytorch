```python
import torch
from matplotlib import pyplot as plt
```

```python
x = torch.arange(-5., 5., 0.1)
```

```python
# take any value and squash it b/w 0 and 1
y = torch.sigmoid(x)
```

```python
plt.plot(x.numpy(), y.numpy())
```

Drawbacks of using sigmoid: the output can reach high values very quickly. It has high values for most of the input values which is a problem. It can result in gradients becoming zero or overflowing the floating point value.

These phenomena are known as **vanishing gradient** or **exploding gradient problem** respectively.

```python
# squash values b/w -1 and 1
y = torch.tanh(x)
```

```python
plt.plot(x.numpy(), y.numpy())
```

ReLU is one of the most important activation functions in deep learning and it is fairly simple. it clips negative values to 0. It's as simple as max(x, 0)

```python
relu = torch.nn.ReLU()
y = relu(x)
```

```python
plt.plot(x.numpy(), y.numpy())
```

Sometimes ReLU can cause outputs of a network to simply become zero and never be revived again. This is known as dying ReLU problem and is solved with variants such as leaky ReLU and Parametric ReLU where the leak coefficient a is learned

max(x, ax)

```python
prelu = torch.nn.PReLU(num_parameters=1)
y = prelu(x)
```

```python
plt.plot(x.numpy(), y.detach().numpy())
```

Softmax is usually used for the output layer of a classification problem since it squashes b/w 0 and 1 and the probabilities add to 1. We use it along with cross entropy loss
